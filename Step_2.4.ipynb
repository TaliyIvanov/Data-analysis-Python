# **Группировка, сортировка и корреляция**

```python
#import frameworks
import pandas as pd
```

```python
#Загрузка наших данных
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/students.csv', delimiter=',')
df.info()
```

```python
#Делаем таблицу, которая состоит не из всех столбцов
df_cut=df[['Age','Growth','Weight']]
df_cut.head()
```

# Сортировка

```python
#Сортировка по убыванию, т.к. стоит False, если не указывать значение "ascending", то по умолчанию указывается True
df_cut.sort_values(by='Age', ascending=False)
```

```python
#Сортировка по убыванию, т.к. стоит True, если не указывать значение "ascending", то по умолчанию указывается True
df_cut.sort_values(by='Age', ascending=True)
```

```python
#Сортируем по нескольким столбцам, если не указывать значение "ascending", то по умолчанию указывается True
df_cut.sort_values(['Age','Growth','Weight'], ascending=[False, False, False])
```

```python
#создаем дополнительную таблицу с отсортированными данными
df_sorted=df_cut.sort_values('Growth', ascending=False)
```

```python
#с помощью функции .iloc[1,] мы можем получать определнную запись из нашей таблицы
df_sorted.iloc[0,]
```

# Коэффициент корреляции
Он нужен для того, чтобы измерять степень зависимости одного признака от другого

Максимальные значения КК лежат в диапазоне: [-1:1].В этом случаем зависимость линейная.

Если КК отрицательный, то между признаками есть обратная зависимость.

Если КК очень близок, то фактически зависимости нет.

Если -0.5 < КК < 0.5, то, вероятнее всего никакой зависимости нет.
