Будем заполнять пропуски в признаке "Вес". Но не просто так. Разобьем нашу таблицу на две таблицы:

females=(все женщины из исходной таблицы)

males=(все мужчины из исходной таблицы)

Далее в таблице females заполните пропуски в столбце "Вес", записав туда средний вес женщин. В таблице males заполните пропуски в столбце "Вес", записав туда средний вес мужчин.

Далее объедините таблицы males и females в одну, запустив код:

df_new=males.append(females)

Далее в таблице df_new подсчитайте средний вес людей. Запишите его в окошко ответа.

```python
#создаем таблицу №1 по категориальному признаку
females=df[df['Sex']== 'женский']
females.head()
```

```python
#создаем таблицу №2 по категориальному признаку
males=df[df['Sex']== 'мужской']
males.head()
```

```python
#Заполняем все пропуски в таблице females средними значениями
females['Weight']=females['Weight'].fillna(females['Weight'].mean())
```

```python
#Заполняем все пропуски в таблице males средними значениями
males['Weight']=males['Weight'].fillna(males['Weight'].mean())
```

```python
#Объединяем таблицы в одну с помощью кода из задания:
df_new=males.append(females)
```

```python
#Считаем средний вес
round(df_new['Weight'].mean(),2)
```

```python
#Интересное решение через группировку данных
grp = df.groupby('Sex')['Weight'].mean()
for sex in grp.index:
    df.loc[(df['Sex'] == sex) & (df['Weight'].isna()), 'Weight'] = grp.loc[sex]
print(f'Средний вес людей = {df.Weight.mean().round(3)}')
```
