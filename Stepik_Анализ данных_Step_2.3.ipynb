**Как делать разбиение таблицы и получать выборку данных?**

(общая схема)

1.Узнать имя столбца с ключевым признаком.

2.Узнать, в каком виде информация хранится в нем. метод .info() 

3.Составить выражение вида(Если в столбце хранятся строки)

df_new = df[df['имя столбца'] == 'значение']

4.В последнем выражении символ ==(проверка на равенство) можно заменить на !=(проверка на неравенство)

5.Если в столбце хранятся числа, то помимо операций ==,!= можно применять операции сравнения >,>=,<,<=. Числа в ковычки заключать не надо.

df_new = df[df['имя столбца с числами'] == 123456789]

```python
#Разбиваем людей на три группы
# Люди с ростом меньше 165см
# Люди с ростом от 165см до 175см
# Люди с ростом больше 175см
df_0_165 = df[df['Growth'] < 165]
df_165_175 = df[(df['Growth'] <= 175) & (df['Growth'] >= 165)]
df_175_8 = df[df['Growth'] > 175]
```

```python
#Посчитаем средний рост для каждой из групп
df_0_165['Growth'].mean(), df_165_175['Growth'].mean(), df_175_8['Growth'].mean()
```

```python
#Найти людей, у которых прошел ДР в момент участия в опросе
# df['Age'] df['Year of birth'] Работаем с этими признаками
#У человека прошел ДР, если сумма года рождения и возраста равна 2022 году.
#С помощью функции shape получаем количество людей.

df[df['Age']+df['Year of birth'] == 2022].shape 
```

```python
#Получим всех людей, у кого рост ниже среднего
df[df['Growth'] < df['Growth'].mean()].shape
```

```python
#Получим людей с максимальным возрастом 
df[df['Age'] == df['Age'].max()]
```

```python
#Получим людей с минимальным возрастом 
df[df['Age'] == df['Age'].min()]
```
