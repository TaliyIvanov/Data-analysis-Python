**Задача регрессии** - предсказание числового признака

Предсказываемый признак называется **целевым**

```python
#import frameworks
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
#Загрузка наших данных
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/students.csv', delimiter=',')
#Создание укороченной таблицы т.к. мы ищем зависимость роста от размера ноги и наоборот
df_cut=df[['Growth', 'Shoe size']]
#Исключаем нулевые значения
df_cut=df_cut.dropna()
```

```python
sns.scatterplot(data=df_cut, x='Shoe size', y='Growth')
```

Гипотеза: рост линейно зависит от размера обуви 
Growth=a*ShSize+b

Коэффициенты a и b на данный момент не известны

Модель регрессии как раз умеется находить оптимальные коэффициенты

```python
LienearRegression = LinearRegression() #Создание "ИИ"
results=LienearRegression.fit(df_cut['Shoe size'].values.reshape(-1,1), y=df_cut['Growth'].values) 
#С помощью .fit() просим подогнать результаты
# Целевой признак - рост. Нецелевой признак - размер обуви.
```

```python
df_cut['Shoe size'].values
```

```python
df_cut['Shoe size'].values.reshape(-1,1)
```

```python
results.coef_, results.intercept_ #Вычисленные коэффициенты
#a=2.7 , b=60.37
```

Прогоним через нашу модель людей и для каждого из них она предскажет рост. И мы сравним предсказанный рост с фактическим ростом.

```python
df_cut['Predicted Growth']=results.predict(df_cut['Shoe size'].values.reshape(-1,1))
```

```python
from sklearn.metrics import mean_absolute_error #данная библиотека подключается для сравнения полученных значений и получения средней ошибки
mean_absolute_error(df_cut['Growth'], df_cut['Predicted Growth']) #Передаем истинный рост и предсказанный рост
```

#Посмотрим разницу в пальцах студентов

```python
df_finger=df[['MIddle and index finger','Middle and ring finger','Middle and little finger']]
```

```python
LienearRegression = LinearRegression() #Создание "ИИ"
results=LienearRegression.fit(df_finger[['Middle and ring finger','Middle and little finger']].values.reshape(-1,2), y=df_finger['MIddle and index finger'].values) 
results.intercept_, results.coef_
```

MIddle and index finger = 0.107*Middle and ring finger+0.04Middle and little finger+7.44

```python
df_finger['Predicted']=results.predict(df_finger[['Middle and ring finger','Middle and little finger']].values.reshape(-1,2))
```

```python
mean_absolute_error(df_finger['Predicted'], df_finger['MIddle and index finger'])
```

#Проверка точности модели на новой выборке
До этого моменты мы тренировали модель и проверяли ее качество (mean_absolute_error) на одних и тех же данных.
Вообще говоря, это не есть корректно. Т.к. модель была построена с участием данных, на которых она потом проверяется.

Поэтому в анализе данных поступают так:

У вас есть две таблицы: тренировочная и тестовая выборка.
Модель строится по тренировочной выборке. Качество ее проверяется по тестовой выборке.
Сейчас мы это реализуем.
