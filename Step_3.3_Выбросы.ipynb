#Первая Процедура поиска выбросов 

m - среднее значение признака .mean()

s = отклонение признака .std()

Построить интервал [m-3s,m+3s]. Все объекты, значение признака которых не попадет в этот интервал, нужно объявить выбросами и удалить из таблицы.

Поищем выбросы относительно признака "BEC"

```python
m=df['Weight'].mean()
s=df['Weight'].std()
m-3*s, m+3*s
```

```python
df=df[(df['Weight']<m+3*s) & (df['Weight']>m-3*s)]
sns.boxplot(df['Weight'])
```

```python
#Все то же самое для возраста
m=df['Age'].mean()
s=df['Age'].std()
m-3*s, m+3*s
```

```python
df=df[(df['Age']<m+3*s) & (df['Age']>m-3*s)]
sns.boxplot(df['Age'])
```

#Вторая процедура поиска выбросов

a - 25ая процентиль признака

b - 75ая процентиль признака

Построим интервал:

[a-1.5(b-a), b+1.5(b-a)]

Значения признака, непопавшего в указаный интервал объясним выбросом.

```python
#Загрузка наших данных т.к. мы ранее изменяли df
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/students.csv', delimiter=',')
```

```python
df.describe()
```

```python
df['Age'].quantile(.25), df['Age'].quantile(.75)
```

```python
a = df['Age'].quantile(.25)
b = df['Age'].quantile(.75)
```

```python
# [a-1.5(b-a), b+1.5(b-a)]
# расчет границ этого интервала
a-1.5*(b-a), b+1.5*(b-a)
```

```python
df=df[(df['Age']<b+1.5*(b-a)) & (df['Age']>a-1.5*(b-a))]
sns.boxplot(df['Age'])
```
