```python
#import frameworks
import pandas as pd
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier #подключение "леса деревьев"
```

```python
#Загрузка наших данных
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/students.csv', delimiter=',')
```

```python
#Создаем таблицу с теми, которые нам необходимы
df_cut=df[['Growth', 'Weight', 'Sex','Hair length', 'Children number']]
df_cut=df_cut.dropna()
```

```python
sns.pairplot(df_cut, hue='Sex')
```

```python
#создаем и запускаем обучаться модель ИИ
model=RandomForestClassifier(max_depth=2, random_state=0) #Указываем другое название модели ИИ
model.fit(df_cut[['Growth', 'Weight','Hair length', 'Children number']].values.reshape(-1,4), y=df_cut['Sex'].values)
```

```python
# загружаем данные тестовой выборки
df_test=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/students_test.csv', delimiter=',')
df_test_cut=df_test[['Growth', 'Weight', 'Sex','Hair length', 'Children number']]
df_test_cut=df_test_cut.dropna() #удаляем данные, имеющие пустые значения
```

```python
result=model.predict_proba(df_test_cut[['Growth', 'Weight','Hair length', 'Children number']].values.reshape(-1,4))
print(result) #вероятности принадлежности классами. Первый столбец - показывается вероятность того, что относится к одному классу. Второй, что относится ко второму классу
```

```python
df_test_cut['pr class 0']=result[:,0]
df_test_cut['pr class 1']=result[:,1]
df_test_cut.head()
```

```python
#Здесь мы можем настраивать порог % значений. которые попадают к тому или иному классу
df_test_cut[(df_test_cut['pr class 1']<0.5) & (df_test_cut['Sex']=='мужской')].head()
```

```python
df_test_cut.sort_values(by='pr class 1', ascending=True).head(50)
```
