```python
#import frameworks
import pandas as pd
import seaborn as sns
from sklearn.ensemble import GradientBoostingClassifier #подключение Бустинга
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/students.csv', delimiter=',')
```

```python
df_cut= df.iloc[:,6:17]
```

```python
model=GradientBoostingClassifier(random_state=0) 
model.fit(df.iloc[:,6:17].values.reshape(-1,11), y=df['Your insitute'].values)
```

```python
# загружаем данные тестовой выборки
df_test=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/students_test.csv', delimiter=',')
```

```python
df_test['Predicted']=model.predict(df_test.iloc[:,6:17].values.reshape(-1,11))
```

```python
pd.crosstab(df_test['Predicted'], df_test['Your insitute'])
```

```python
from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(df_test['Predicted'], df_test['Your insitute'])
```
