# Кластеризация

Вам нужно автоматически разбить данные на несколько групп  (кластеров) так, чтобы каждая группа состояла из похожих друг на друга объектов. Критерий похожести ИИ находит сам.

```python
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/задания к курсу/students.csv', delimiter=',')
df_cut=df[['Weight','Growth','Sex']]
```

```python
df_cut=df_cut.dropna()# показать существенность этой строки
```

```python
kmeans = KMeans(n_clusters=2, random_state=1000)
kmeans.fit(df_cut[['Weight','Growth']])
df_cut['Label']=kmeans.labels_
```

```python
sns.scatterplot(data=df_cut, x='Weight', y='Growth', hue='Label')
```

```python
sns.scatterplot(data=df_cut, x='Weight', y='Growth', hue='Label')
```

```python
sns.scatterplot(data=df_cut, x='Weight', y='Growth', hue='Sex')
```

Быстрый подсчет объектов, удовлетворяющих тому или иному условию

```python
sum((df_cut['Sex']=='мужской')& (df_cut['Label']==1))
# будет подсчитано количество объектов удовлетворящих условию в скобках
```

```python
sum((df_cut['Sex']=='женский') & (df_cut['Label']==0))
```

# Следующий код не предназначен для людей младше 18 лет.

```python
kmeans = KMeans(n_clusters=3, random_state=666)
kmeans.fit(df_cut[['Weight','Growth']])
df_cut['Label']=kmeans.labels_
```

```python
sns.scatterplot(data=df_cut, x='Weight', y='Growth', hue='Label')
```

```python
sns.scatterplot(data=df_cut, x='Weight', y='Growth', hue='Label')
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=42) # 2 компоненты, значит 2-мерное представление

X_reduced = pca.fit_transform(df.dropna()[['Growth', 'Weight', 'Hair length', 'Age']])
```

```python
sns.scatterplot(X_reduced[:,0], X_reduced[:,1], hue=df.dropna()['Sex'])
```

```python
df_cut=df[['Shoe size', 'Weight']]
df_cut=df_cut.dropna()
kmeans = KMeans(n_clusters=2, random_state=1)
kmeans.fit(df_cut)
df_cut['Label']=kmeans.labels_
```
